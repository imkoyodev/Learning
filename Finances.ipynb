## 5. Construcción de Portafolio de Inversión

La selección de productos de inversión y su distribución porcentual en el portafolio de inversión se desarrollará empleando el modelo de Markowitz. Posteriormente, se efectuará la optimización del portafolio empleando la metodología propuesta por Kolm et al (2014).

```python
# Librerias Empleadas
from pypfopt import risk_models, expected_returns, plotting, EfficientFrontier, objective_functions, DiscreteAllocation
import cvxpy as cp
import matplotlib.pyplot as plt 
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# Descarga de Información y almacenamiento en Parquet

WaterUtilitiesTickets = ['CWCO', 'AWK', 'WTRG', 'MSEX', 'SJW']
ElectricUtilitiesTickets = ['ED', 'EXC', 'AEP', 'DUK', 'SO']
FinancialSectorTickets = ['TD', 'WFC', 'BAC', 'JPM', 'HSBC']
EnergySectorTickets = ['CVX', 'WMB', 'BEP', 'NRG', 'SLB']
TechnologySectorTickets = ['IBM', 'CSCO', 'SATS', 'FTNT', 'SNPS', 'PANW', 'MU']
TreasuryBondsTickets = ['VGLT', 'VGIT', 'UBT', 'UST']
SP500Tickets = ['SPY', 'VOO']
Tickets = WaterUtilitiesTickets + ElectricUtilitiesTickets + EnergySectorTickets + TechnologySectorTickets + TreasuryBondsTickets + SP500Tickets          

HistoricalData = yf.download(Tickets, start='2015-02-28', end='2025-03-01')
HistoricalData.to_parquet('HistoricalData.parquet')
```

```python
# Carga de archivo Parquet
HistoricalData = pd.read_parquet('HistoricalData.parquet')
HistoricalData = HistoricalData['Close'].dropna(axis=1)
HistoricalData = HistoricalData.drop(SP500Tickets, axis=1)
```

```python
# Retornos experados
mu = expected_returns.capm_return(HistoricalData)
```

A partir de la estimación de la rentabilidad esperada para los activos en análisis, se observa que 9 de los 26 en análisis han presentado una rentabilidad inferior a la ofrecida por los ETF que replican el comportamiento del índice S&P 500. Destacándose la baja rentabilidad de los activos asociados a los bonos del tesoro. Igualmente, algunos activos vinculados al sector de servicios públicos presentan una rentabilidad ligeramente menor a la ofrecida por el S&P 500, situación esperada debido a la naturaleza del activo. Con el objetivo de reducir el impacto negativo de estos activos de bajo rendimiento, se eliminarán todos aquellos que presenten una rentabilidad inferior a 0.124, reduciendo el número de activos de 26 a 19. 

```python
ax = mu.sort_values(ascending=False).plot.bar(alpha=0.4, title='Retornos Esperados');
ax.axhline(y=0.127954, color='red')
ax.text(x= 17, y=0.128954, s='Rentabilidad S&P 500');
```

```python
HistoricalData = HistoricalData.drop(mu[ mu.values < 0.124].index.values, axis=1)
```

```python
## Matriz de Covarianzas - Ledoit-Wolf shrinkage
S = risk_models.CovarianceShrinkage(HistoricalData,).ledoit_wolf()

# Retornos experados
mu = expected_returns.capm_return(HistoricalData)
```

A partir de la matriz de covarianzas, se puede observar que los 19 activos en análisis han presentado una relación positiva en los últimos 10 años. En principio, este comportamiento puede afectar el desempeño del portafolio, debido a la carencia de activos que presenten un comportamiento complementario y permitan compensar momentos de reducción de ganancias por situaciones estacionales. 

```python
plotting.plot_covariance(S, plot_correlation=True);
```

A partir de la Barrera Logarítmica descrita en Kolm et al. (2014), construimos un portafolio de inversión optimizado. Como resultado, se espera una rentabilidad anual del 15.4%, con una volatilidad del 18.4% y un Sharpe Ratio de 0.84. Estos valores son coherentes con lo reportado para portafolios con un horizonte de inversión de largo plazo. En la imagen que se observa más adelante, se presenta la distribución del portafolio en cada una de las acciones. 

```python
# Funcion de optimización propuesta en Kolm et al (2014):
def logarithmic_barrier_objective(w, cov_matrix, k=0.1):
    log_sum = cp.sum(cp.log(w))
    var = cp.quad_form(w, cov_matrix)
    return var - k * log_sum
```

```python
tickers = list( HistoricalData.columns )
initial_weights = np.array([1/len(tickers)] * len(tickers))
ef = EfficientFrontier(mu, S)

# 1% Comision del intermediario o broker
ef = EfficientFrontier(mu, S, weight_bounds=(0.01, 0.2))
ef.convex_objective(logarithmic_barrier_objective, cov_matrix=S, k=0.01)
weights = ef.clean_weights()
ef.portfolio_performance(verbose=True);
```

```python
portfolio_value = 30000
cleaned_weights = ef.clean_weights()
latest_prices = HistoricalData.loc['2025-02-28 00:00:00']
da = DiscreteAllocation(cleaned_weights, latest_prices, total_portfolio_value=portfolio_value)
```

```python
allocation, leftover = da.greedy_portfolio()
initial_cost = HistoricalData.loc['2025-02-28 00:00:00'].mul( pd.Series(allocation), axis=0 ).dropna().sum()
print("Fondos Disponibles: USD {:.2f}".format(leftover))
print("Valor de la Inversión: USD {:.2f}".format( initial_cost ))
```

```python
ax = pd.Series(allocation).sort_values(ascending=False).plot.bar(figsize=(10,15), 
                                    ylabel="Numero de Assets", 
                                    xlabel="Asset Ticker",
                                    title="Conformación del Portafolio",
                                    alpha=0.4)
```
